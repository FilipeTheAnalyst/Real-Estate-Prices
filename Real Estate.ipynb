```python
import pandas as pd
from bs4 import BeautifulSoup
import time
from selenium.webdriver.chrome.options import Options
from selenium import webdriver
import os
import re
```

## Web Scraping

```python
houses_url = []
for i in range(1, 16):
    page_number = str(i)
    test_url = f"https://casa.sapo.pt/comprar-apartamentos/t3/seixal/?pn={page_number}"

    pwd = os.getcwd()
    driver_path = pwd + "/chromedriver.exe"
    wd = webdriver.Chrome(executable_path=driver_path)

    wd.get(test_url)

    time.sleep(1)

    with open(f"house_pages/{page_number}.html", "w", encoding='utf-8') as file:
        file.write(wd.page_source)

    with open(f"house_pages/{page_number}.html", 'r', encoding='utf8') as reader:
        html_string = reader.read()
        soup = BeautifulSoup(html_string, 'html.parser')
        for tag in soup.find_all(class_="swiper-wrapper"):
            if re.findall(r"https:\W+casa\S+[^']", tag.find('a')['href']):
                houses_url.append(re.findall(r"https:\W+casa\S+[^']", tag.find('a')['href']))
            else:
                houses_url.append(['https://casa.sapo.pt'+ tag.find('a')['href']])
```

## Parsing Data

```python
for i in range(len(houses_url)):
    pwd = os.getcwd()
    driver_path = pwd + "/chromedriver.exe"
    wd = webdriver.Chrome(executable_path=driver_path)
    wd.get(houses_url[i][0])

    time.sleep(1)

    with open(f"house_prices/house_{i}.html", "w", encoding='utf-8') as file:
        file.write(wd.page_source)
```

## Listing all files to iterate through

```python
files = os.listdir("house_prices/")
files.sort(key=lambda f: int(re.sub('\D', '', f)))
files
```

```python
prices = []
estado = []
area_util = []
area_bruta = []
ano_construcao = []
description = []
location = []
url = []
casas_banho = []

for i in files:
    aux = 0
    size = 0
    print(i)
    with open(f"house_prices/{i}", 'r', encoding='utf8') as reader:
        html_string = reader.read()
        soup = BeautifulSoup(html_string, 'html.parser')
        prices.append(soup.find(class_='detail-title-price-value').text.strip())
        try:
            area_bruta.append(soup.find(class_='detail-main-features-item-title', text=re.compile('Área bruta')).find_next(class_='detail-main-features-item-value').text.strip())
        except AttributeError as e:
            area_bruta.append('') 
        try:
            area_util.append(soup.find(class_='detail-main-features-item-title', text=re.compile('Área útil')).find_next(class_='detail-main-features-item-value').text.strip())
        except AttributeError as e:
            area_util.append('') 
        try:
            ano_construcao.append(soup.find(class_='detail-main-features-item-title', text=re.compile('Ano de construção')).find_next(class_='detail-main-features-item-value').text.strip())
        except AttributeError as e:
            ano_construcao.append('')
        try:
            estado.append(soup.find(class_='detail-main-features-item-title', text=re.compile('Estado')).find_next(class_='detail-main-features-item-value').text.strip())
        except AttributeError as e:
            estado.append('N/A')
        description.append(soup.find(class_='detail-description-text').text.strip())
        location.append(soup.find(class_='detail-title-location').text.strip())
        url.append(re.findall(r".+html", soup.find(rel='amphtml')['href'])[0])
        for item in soup.find_all(class_='detail-features-item'):
            size+=1
            if re.findall(r"\w+\W\w\W\s\w{2}\sBanho", item.text.strip()) and aux == 0 :
                casas_banho.append(item.find('strong').text.strip())
                aux = 1
            elif aux == 0 and size == len(soup.find_all(class_='detail-features-item')):
                casas_banho.append('')
                        
```

```python
df = pd.DataFrame(data={"Preco":prices, "Area Bruta":area_bruta, "Area Util":area_util, "Estado":estado,
                    "Localizacao":location, "Link":url, "Ano de Construção":ano_construcao, "Casa(s) de Banho":casas_banho, "Descricao":description})
df
```

```python
df['Preco'] = df['Preco'].str.replace('€', '', regex=True)
df['Preco'] = df['Preco'].str.replace('.', '', regex=True)
df['Preco'] = pd.to_numeric(df['Preco'])
```

```python
df['Area Util'] = df['Area Util'].str.replace('m²', '', regex=True)
df['Area Util'] = df['Area Util'].str.replace(' ', '', regex=True)
df['Area Bruta'] = df['Area Bruta'].str.replace('m²', '', regex=True)
df['Area Bruta'] = df['Area Bruta'].str.replace(' ', '', regex=True)
```

```python
df['Ano de Construção'] = pd.to_numeric(df['Ano de Construção'])
df['Area Util'] = pd.to_numeric(df['Area Util'])
df['Area Bruta'] = pd.to_numeric(df['Area Bruta'])
```

```python
df['Casa(s) de Banho'] = pd.to_numeric(df['Casa(s) de Banho'])
```

```python
df.dtypes
```

```python
df = df.dropna(subset=['Preco'])
df
```

### Fill value in Ano de Construção with the present year when Estado = 'Novo' and Ano de Construção is NA

```python
df.loc[(df['Estado'] == 'Novo') & (df['Ano de Construção'].isna()), 'Ano de Construção'] = 2022
```
